```python
import pandas as pd
import plotly.express as px
```

```python
# read dados
df = pd.read_excel ("/content/val kickoff 2024.xlsx")

df.head()
```

```python
#drop columns
df = df.drop(columns=["A/Map","D/Map", "K/Map"])
```

```python
df.head()
```

```python
df.info()
df.isna().sum()
```

```python
#if you had Nan
df.dropna()
```

```python
a = df["Agents"].value_counts()
print(a)

df["Agents"].value_counts(normalize=True).map("{:.1%}".format)

# 55 players
# Around 58.2%(32) of players used at least 3 agents during the championship.
# 32.7%(18) of players used at least 2 agents and 9.1% (5) used 1 agent
```

```python
plt = px.histogram(df, x="Agents", color= "Player",
                   title = "Number of agents used by players (1 to 3)",
                    color_discrete_sequence= px.colors.qualitative.Plotly)

plt.update_layout(
    title_font_family="Berlin Sans FB",
    title_font_color = "#ff8c00",
    font_color = "black"
)

plt.show()
```

```python
df.head()
```

```python
df_kda = df[["Player","KDA", "Maps"]]
df_kda = df_kda.sort_values(by=["KDA"],ascending = False)
df_kda = df_kda.head(10)
df_kda

# the top kda playing in loud and played 13 maps compared tenz who playing in sentinels and
#winning the vct americas kickoff and played 21 maps
```

```python
plt = px.bar(df_kda, x = "Player", y = "KDA", color = "Maps", text_auto=True,
            title = "Top 10 KDA VCT 2024: Americas Kickoff ")

plt.update_layout(
    font_family="Berlin Sans FB",
    title_font_color = "#ff8c00",
    font_color = "black"


)
plt.show()
```

```python
df_kills = df[["Player", "K", "Maps"]]
df_kills = df_kills.sort_values(by = ["K"],ascending = False)
df_kills = df_kills.head(10)
df_kills

# The top 5 with the most kills are from the Americas Kickoff champion team
```

```python
plt = px.bar (df_kills , x = "Player", y = "K" , color = "Maps",
              text_auto = True,
              title = "Top 10 Most Kills VCT 2024: Americas KickoffF")

plt.update_layout(
    font_family="Berlin Sans FB",
    title_font_color = "#ff8c00",
    font_color = "black"
)
plt.show()
```
