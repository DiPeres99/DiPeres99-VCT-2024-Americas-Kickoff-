```python
import pandas as pd
```

```python
df = pd.read_excel("kdatimesatt.xlsx")
df.head()
```

```python
df.info()
df.isna().sum()
```

```python
df1 = df[["times","kda/time", "maps"]]
df1 = df1.sort_values(by=["kda/time"], ascending = False)
df1.head()
```

```python
import plotly.express as px
```

```python
plt = px.bar(df1, x = "times", y = "kda/time",text_auto=True,
                 color ="times",
                 title = "KDA by Teams VCT 2024: Americas Kickoff")

plt.update_layout(
    font_family="Berlin Sans FB",
    title_font_color = "#ff8c00",
    font_color = "black"
)

plt.show()
```
